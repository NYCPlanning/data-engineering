```python
cdta = df.loc[:,['cdta', 'cdta_name']]\
        .drop_duplicates()\
        .rename(columns={'cdta':'geoid', 'cdta_name': 'geogname'})\
        .assign(geotype='CDTA2020')
cdta
```

```python
boro = df.loc[:,['borocode', 'boro']]\
        .drop_duplicates()\
        .rename(columns={'borocode':'geoid', 'boro': 'geogname'})\
        .assign(geotype='Boro2020')
boro
```
